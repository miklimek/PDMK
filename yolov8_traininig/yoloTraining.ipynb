Importowany jest model YOLOv8 nano, już wytrenowany na zbiorze COCO. Zbiór treningowy ustawiany jest na zbiór ping-pang2.0 (plik data.yaml zawiera konfigurację).
Długość treninigu ustawiana jest na 100 epok, włączone jest zapisywanie wag co 5 epok, treninig uruchamiany jest na GPU

```python
from ultralytics import YOLO

# Load a model
model = YOLO('yolov8n.pt')  # load a pretrained model (recommended for training)

# Train the model with GPU
model.train(data='rackets/data.yaml', epochs=100, save=True, save_period=5, workers=3, imgsz=640, device=0)
```

```python
from PIL import Image
from ultralytics import YOLO

model = YOLO('runs/detect/train15/weights/best.pt') # załadowanie modelu z najlepszymi wytrenowanymi wagami
results = model(['rackets/test/test.png']) # detekcja na obrazie

for r in results:
    im_array = r.plot()  # wyświetlenie wyników detekcji
    im = Image.fromarray(im_array[..., ::-1]) 
    im.show()
```
